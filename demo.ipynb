# faker-ai-provider Demo

Interactive demonstration of the faker-ai-provider library for generating realistic AI/ML test data.

## Setup

```python
from faker import Faker
from faker_ai import AiProvider

fake = Faker()
fake.add_provider(AiProvider)
print("✅ Provider loaded!")
```

## Basic Data Generation

```python
print("AI Models:")
for _ in range(5):
    print(f"  • {fake.ai_model()}")

print("\nAI Companies:")
for _ in range(5):
    print(f"  • {fake.ai_company()}")
```

## Correlated Data

Models are linked to their companies, architectures, and capabilities.

```python
companies = ["OpenAI", "Anthropic", "Google DeepMind", "Meta AI"]

for company in companies:
    model = fake.ai_model_for_company(company)
    print(f"{company}: {model}")
```

## Complete Model Scenarios

```python
for i in range(3):
    scenario = fake.model_scenario()
    print(f"\nScenario {i+1}:")
    print(f"  Model: {scenario['model']}")
    print(f"  Company: {scenario['company']}")
    print(f"  Architecture: {scenario['architecture']}")
    print(f"  Parameters: {scenario['parameters']}")
    print(f"  Tasks: {', '.join(scenario['tasks'][:3])}")
```

## Composite Methods

```python
print("Full Model Specs:")
for _ in range(3):
    print(f"  • {fake.full_ai_model_spec()}")

print("\nTraining Run:")
run = fake.ai_training_run()
for k, v in run.items():
    print(f"  {k}: {v}")

print("\nExperiment:")
exp = fake.ai_experiment()
for k, v in exp.items():
    print(f"  {k}: {v}")
```

## Seeding for Reproducibility

```python
fake.seed_instance(42)
print("With seed 42:")
print(f"  Model: {fake.ai_model()}")
print(f"  Company: {fake.ai_company()}")

fake.seed_instance(42)
print("\nWith seed 42 again (same results):")
print(f"  Model: {fake.ai_model()}")
print(f"  Company: {fake.ai_company()}")
```
